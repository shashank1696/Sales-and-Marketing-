```python
data['History'] = data['History'].fillna('new')
#replace na values with new
```

```python
data['History'].head(20)
```

```python
data = pd.get_dummies(data)
data
```

```python
data = data.drop(['History_new'],axis=1)
```

```python
data.head(10)
```

```python
y = data['target']
x=data.drop('target',axis = 1)
```

```python
##create train and test parts of the data
import sklearn.model_selection as model_selection
```

```python
x_train,x_test,y_train,y_test = model_selection.train_test_split(x,y,test_size=0.2,random_state = 200)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
import sklearn.tree as tree
```

```python
clf = tree.DecisionTreeClassifier(max_depth = 3)
```

```python
clf.fit(x_train,y_train)
```

```python
##create a graphical representation of the tree model
import pydotplus
import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin'
#os.chdir(dt_dir)
```

```python
graph = tree.export_graphviz(clf,feature_names = x_train.columns,class_names=["0","1"],proportion = True,out_file = None,filled = True)
```
